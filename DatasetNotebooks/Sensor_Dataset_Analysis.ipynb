```python
scaler = StandardScaler()

# Calcola le statistiche di normalizzazione sul dataset di addestramento e applicale
X_train_normalized = scaler.fit_transform(X_train)

# Applica la stessa normalizzazione al dataset di test
X_test_normalized = scaler.transform(X_test)
```
